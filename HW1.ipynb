```python
# 想解決的問題: 了解內向人格的同學參與社團積極度
import pandas as pd

# read csv file
shy = pd.read_csv("shy.csv")
frequently = pd.read_csv("frequency.csv")
```

```python
frequency = list(frequently['frequency'])
personality = list(shy['personality'])
set_frequency = set(frequently['frequency'])
set_personality = set(shy['personality'])
set_frequency.remove('FALSE')
set_personality.remove('FALSE')
```

```python
# 外向且積極參與社團活動的人的ID
# 差集
set(frequency).difference(set(personality))
```

```python
# 內向且無積極參與社團活動的人的ID
# 差集(第二種寫法)
print(set_personality-set_frequency)
print("內向且無積極參與社團活動的人數:",len(set_personality-set_frequency))
```

```python
# 內向且積極參與社團活動的人的ID
# 交集
print(set_frequency & set_personality)
print("內向且積極參與社團活動的人數:",len(set_frequency & set_personality))
```

```python
# 所有人的ID
# 聯集
print(set_frequency |set_personality)
```

```python
# 外向且無積極參與社團活動的人的ID
# 對稱差集
set(frequency).symmetric_difference(set(personality))
```
